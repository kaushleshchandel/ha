## Web scraping departure times from metlink.org

This is an example to get the required data to set up an the [`scrape` sensor](https://home-assistant.io/components/sensor.scrape/).

Importing the needed modules.

```python
import requests
from bs4 import BeautifulSoup
```

The URL for the departure times are located that `/departures`.

```python
URL = 'https://www.metlink.org.nz/stop/7730/departures'
```

With `requests` the website is retrieved and with `BeautifulSoup` parsed.

```python
raw_html = requests.get(URL).text
data = BeautifulSoup(raw_html, 'html.parser')
print(data)
```

Using `span` as selector as mentioned at [CSS selectors](https://www.crummy.com/software/BeautifulSoup/bs4/doc/#css-selectors) the departure time can be retrieved.

```python
print(data.select('span:nth-of-type(3)')[0].text)
```
